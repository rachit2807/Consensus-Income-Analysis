### Missing Values

```python
missing_df =  df.isnull().sum().to_frame().rename(columns={0:"Total No. of Missing Values"})
missing_df["% of Missing Values"] = round((missing_df["Total No. of Missing Values"]/len( df))*100,2)
missing_df
```

```python
numerical_vars = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_vars = df.select_dtypes(include=['object']).columns.tolist()                           
print('Numerical variables:', numerical_vars)
print('Categorical variables:', categorical_vars)
```

```python
categorical_count = df.select_dtypes(include='object').shape[1]
numerical_count = df.select_dtypes(exclude='object').shape[1]

print(f"Number of categorical variables: {categorical_count}")
print(f"Number of numerical variables: {numerical_count}")
```

```python
numerical_columns = df.select_dtypes(include=np.number).columns.tolist()
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())
```

```python
categorical_columns = df.select_dtypes(exclude=np.number).columns.tolist()
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
df.isnull().sum()
```

### Duplicate Values

```python
df[df.duplicated(keep=False)]
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

cat_unique_counts = df[cat_cols].nunique().reset_index().rename(columns={'index': 'column', 0: 'unique_count'})

cat_non_unique_counts = df[cat_cols].apply(lambda x: x.shape[0] - x.nunique()).reset_index().rename(columns={'index': 'column', 0: 'non_unique_count'})

cat_counts = pd.merge(cat_unique_counts, cat_non_unique_counts, on='column')
print("\nCounts of Unique and Non-Unique Values for Categorical Variables:")
print(cat_counts)
```

# Data Analysis

### Distribution of target variable

```python
sns.countplot(df['income_bracket'])
plt.title('Distribution of Income')
plt.show()
```

#### The dataset contains a relatively balanced distribution of income levels, with a significant portion of individuals earning <=50K.

### Distribution of Age

```python
plt.figure(figsize=(10,6))
sns.histplot(df['AAGE'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#### The age distribution is mostly concentrated between 20 to 60 years, with a peak around 40 years.

### Scatter plot for age vs. hourly pay rate

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='AAGE', y='AHRSPAY', hue='income_bracket', palette='coolwarm')
plt.title('Age vs. Hourly Pay Rate')
plt.xlabel('Age')
plt.ylabel('Hourly Pay Rate')
plt.show()
```

#### There is a positive correlation between age and hourly pay rate, indicating that older individuals tend to have higher hourly pay rates.
